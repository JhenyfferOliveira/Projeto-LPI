```python
import manipulate_data

# Página inicial

print('-'*20, 'AdaFood', '-'*20)

opcao = -1
while opcao not in[0, 1, 2, 3, 4, 5]:
  opcao = int(input(f'\n{1} Novo Restaurante\n{2} Novo cliente\n{3} Restaurantes Existentes\n{4} Clientes Existentes\n{5} Consultar Restaurantes próximos\n{0} Sair\n'))
  if opcao == 1:
    # CADASTRO RESTAURANTE
    # Carrega dados antes de imprimir
    lista_restaurantes = manipulate_data.load_json_file('Cadastro', 'Restaurantes')

    # Verifica se lista de restaurantes existe
    if lista_restaurantes is None:
      lista_restaurantes = []

    # Inputs de cadastro do restaurante
    n_pedidos_restaurante = 0
    print('')
    print('Faça o seu cadastro.')
    nome_restaurante = str(input('Insira o nome do Restaurante:'))
    lat_restaurante = float(input('Insira a latitude: '))
    long_restaurante = float(input('Insira a longitude: '))
    localizacao_restaurante = lat_restaurante, long_restaurante
    print('Cadastro realizado!\n')

    # Adiciona inputs de cadastro à lista restaurante
    restaurante = []
    restaurante.append(nome_restaurante)
    restaurante.append(localizacao_restaurante)
    restaurante.append(n_pedidos_restaurante)
    lista_restaurantes.append(restaurante)

    # Indica itens da coluna + salva a lista de restaurantes cadastrado
    coluna_restaurantes = ['nome_restaurante', 'localizacao_restaurante', 'n_pedidos_restaurante']
    manipulate_data.save_json_file('Cadastro', 'Restaurantes', lista_restaurantes, coluna_restaurantes)


    # CARDÁPIO
    # Carrega dados antes de imprimir
    lista_cardapio = manipulate_data.load_json_file('Cadastro', 'Cardápios')

    # Verifica se lista de cardápios existe
    if lista_cardapio is None:
      lista_cardapio = []

    # Inputs de cadastro do cardápio
    print('Crie seu cardápio.')
    valor_jap = float(input('Insira o valor médio da comida japonesa: '))
    valor_hamb = float(input('Insira o valor médio dos hambúrgueres: '))
    valor_pizza = float(input('Insira o valor médio das pizzas: '))
    print('Cadastro realizado!\n')

    # Adiciona inputs de cadastro ao cardápio
    cardapio = []
    cardapio.append(valor_jap)
    cardapio.append(valor_hamb)
    cardapio.append(valor_pizza)
    lista_cardapio.append(cardapio)

    # Imprime nome do restaurante cadastrado + cardápio
    print('-'*30, 'Restaurante e Cardápio', '-'*30)
    print(f'Informações do Restaurante\nNome: {nome_restaurante}\nLocalização: {localizacao_restaurante}\nNúmero de pedidos: {n_pedidos_restaurante}')
    print('')
    print(f'Cardápio {nome_restaurante}\n{cardapio}\n')

    # Indica itens da coluna + salva a lista de cardápio cadastrado
    coluna_cardapios = ['valor japonesa', 'valor hamburger', 'valor pizza']
    manipulate_data.save_json_file('Cadastro', 'Cardápio', lista_cardapio, coluna_cardapios)



  elif opcao == 2:
    # CADASTRO CLIENTE
    # Carrega dados antes de imprimir
    lista_clientes = manipulate_data.load_json_file('Cadastro', 'Clientes')

    # Verifica se lista clientes existe
    if lista_clientes is None:
      lista_clientes = []

    # Inputs de cadastro do cliente
    print('')
    n_pedidos_cliente = 0
    print('Faça o seu cadastro.')
    nome_cliente = str(input('Insira o seu nome: '))
    lat_cliente = float(input('Insira a latitude: '))
    long_cliente = float(input('Insira a longitude: '))
    localizacao_cliente = lat_cliente, long_cliente
    print('Cadastro realizado!\n')

    # Imprime cliente cadastrado
    print('-'*20, 'Cliente', '-'*20)
    print(f'Informações do Cliente:\nNome: {nome_cliente}\nLocalização: {localizacao_cliente}\nNúmero de pedidos: {n_pedidos_cliente}')
    print('')

    # Adiciona inputs de cadastro à lista restaurante
    cliente = []
    cliente.append(nome_cliente)
    cliente.append(localizacao_cliente)
    lista_clientes.append(cliente)

    # Indica itens da coluna + salva a lista de cliente cadastrado
    coluna_clientes = ['nome_restaurante', 'localizacao_restaurante', 'n_pedidos_restaurante']
    manipulate_data.save_json_file('Cadastro', 'Clientes', lista_clientes, coluna_clientes)



  elif opcao == 3:
    # Carrega dados de restaurantes cadastrados
    lista_restaurantes = manipulate_data.load_json_file('Cadastro', 'Restaurantes')

    # Confere se lista existe
    if lista_restaurantes is None:
        lista_restaurantes = []

    # Imprime dados que estão salvos no caminho indicado
    print('-'*20, 'Restaurantes Cadastrados', '-'*20)

    # lista_restaurantes é uma lista de listas
    for restaurante in lista_restaurantes:
        # 'for restaurante' acessa cada lista pelos índices indicados
        nome = restaurante[0]
        localizacao = restaurante[1]
        n_pedidos = restaurante[2]

        # Imprime os dados
        print(f'Nome: {nome}')
        print(f'Localização: {localizacao}')
        print(f'Número de Pedidos: {n_pedidos}')
        print('-' * 70)



  elif opcao == 4:
    # Carrega dados de clientes cadastrados
    lista_clientes = manipulate_data.load_json_file('Cadastro', 'Clientes')

    # Confere se lista existe
    if lista_clientes is None:
        lista_clientes = []

    # Imprime dados que estão salvos no caminho indicado
    print('-'*23, 'Clientes Cadastrados', '-'*23)

    # lista_clientes é uma lista de listas
    for cliente in lista_clientes:
        # 'for cliente' acessa cada lista pelos índices indicados
        nome = cliente[0]
        nome = cliente[1]

        # Imprime os dados
        print(f'Nome: {nome}')
        print(f'Localização: {localizacao}')
        print('-' * 70)



  elif opcao == 5:
    # Chama função que vai fazer cálculo de latitude e longitude por meio de inputs
    verifica_localizacao()



  elif opcao == 0:
    # Opção de sair do código
    print('\nFim do programa.')
    break



  else:
    # Contrle de erro
    print('Opção inválida. Insira de 1 a 4 ou 0 para sair.')
```

Função cálculo de latitude e longitude

```python
#Nessa fica difinido a localizaçao do restaurante

from geopy.distance import geodesic

# Cria função que faz a verificação da localização
def verifica_localizacao():
  # Coordenadas do restaurante
  restaurante_latitude = float(input("Digite a latitude do restaurante: "))
  restaurante_longitude = float(input("Digite a longitude do restaurante: "))

  # Coordenadas da localização a ser verificada
  localizacao_latitude = float(input("Digite a latitude da localização a ser verificada: "))
  localizacao_longitude = float(input("Digite a longitude da localização a ser verificada: "))

  raio = 3  # Defina o raio desejado em quilômetros

  # Calcula a distância entre o restaurante e a localização
  distancia = geodesic((restaurante_latitude, restaurante_longitude), (localizacao_latitude, localizacao_longitude)).kilometers

  # Verifica se a localização está dentro do raio do restaurante
  if distancia <= raio:
      print(f"A localização está dentro do raio de {raio}km do restaurante.")
  else:
      print(f"A localização está fora do raio de {raio}km do restaurante.")
```
